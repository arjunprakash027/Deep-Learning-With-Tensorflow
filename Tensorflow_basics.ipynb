### Constant and varibale tensor

```python
const = tf.constant([[1,2],[3,4]])
print(const)

print("-------")

variable = tf.Variable([[1,2],[3,4]])
print(variable)
```

### Row concatanation

```python
row_concatanated_tensor = tf.concat(values=[const,variable],axis=0)
print(row_concatanated_tensor)

print("-------")

column_concatanated_tensor = tf.concat(values=[const,variable],axis=1)
print(column_concatanated_tensor)
```

### Zeros and ones tensor

```python
zero_tensor = tf.zeros(shape=[64,64],dtype=tf.int32)
print(zero_tensor)

ones_tensor = tf.ones(shape=[2,8],dtype=tf.int32)
print(ones_tensor)

ones_float_tensor = tf.ones(shape=[64,64],dtype=tf.float32)
print(ones_float_tensor)
```

### Tensor operations

#### Transpose

```python
tensor_transpose = tf.transpose(const)
print(tensor_transpose)
```

#### Type conversion

```python
tensor_transpose_float = tf.cast(tensor_transpose,tf.float32)
print(tensor_transpose_float)
```

#### MUltiplication and matrix multiplication

```python
multiplied_tensor = tf.multiply(tensor_transpose,const)
print(multiplied_tensor)

mat_multiplied_tensor = tf.matmul(multiplied_tensor,const)
print(mat_multiplied_tensor)
```

#### Determinant

```python
mat_multiplied_tensor = tf.cast(mat_multiplied_tensor,tf.float32)#determinant does not work on int
print(mat_multiplied_tensor)
determinant = tf.linalg.det(mat_multiplied_tensor)
print(determinant)
```

### Identity

```python
identity = tf.eye(num_rows=16,num_columns=16,dtype=tf.float32)

print(identity)
```

### Reshape

```python
reshaped_matrix = tf.reshape(tensor=identity,shape=[1,256])
print(reshaped_matrix)
```
