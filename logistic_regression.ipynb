# Logistic regression

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
data = load_iris()
```

```python
data["target_names"]
```

```python
data["feature_names"]
```

```python
x = data["data"]
y = data["target"]
```

```python
print(x)
print(y)
```

```python
dataset = pd.DataFrame(data=np.concatenate((x,y.reshape(-1,1)),axis=1),
columns= ["sl","sw","pl","pw","target"])
```

```python
dataset
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,shuffle=True)
```

```python
x_train
```

```python
input = tf.keras.Input(shape=4)
X = tf.keras.layers.Dense(3,activation=tf.keras.activations.sigmoid)(input)
```

```python
model = tf.keras.models.Model(input,X)
```

```python
from keras.api._v2.keras import metrics

model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.1),
loss = tf.keras.losses.sparse_categorical_crossentropy,metrics=["accuracy"])
```

```python
train = model.fit(x_train,y_train.reshape(-1),
                  validation_data = (x_test,y_test.reshape(-1)),epochs=400)
```

```python
new_ata = [4.5,6.7,3.4,6.7]
y_pred = model.predict([new_ata])
```
